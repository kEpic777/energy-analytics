```python
df1 = pd.read_csv(
    "../data/raw/Imbalance_price_20251101_20251231_Interval.csv",
    sep=";",
    encoding="utf-8-sig",
    decimal=",",
    thousands=".",
    index_col=False
)

print(df1.shape)
print(df1.dtypes)
print(df1.head())
```

```python
df2 = pd.read_csv(
    "../data/raw/Imbalance_price_20260101_20260228_Interval.csv",
    sep=";",
    encoding="utf-8-sig",
    decimal=",",
    thousands=".",
    index_col=False
)

print(df2.shape)
print(df2.dtypes)
print(df2.head())
```

```python
# Concat
df = pd.concat([df1, df2], ignore_index=True)

# Rename columns
df = df.rename(columns={
    "Date": "timestamp",
    "Cpoz [EUR/MWh]": "imbalance_price_pos",
    "Cneg [EUR/MWh]": "imbalance_price_neg",
    "Sipx [EUR/MWh]": "sipx",
    "BO": "settlement_type"
})

# Parse timestamp
df["timestamp"] = pd.to_datetime(df["timestamp"], dayfirst=True)

# Sort
df = df.sort_values("timestamp").reset_index(drop=True)

# Save
df.to_csv("../data/curated/imbalance_si_clean.csv", index=False)

print(df.shape)
print(df.dtypes)
print(df.head())
print(df["timestamp"].min(), df["timestamp"].max())
```

```python
import pandas as pd

months = [
    "../data/raw/Balancing volumes and costs_november2025.csv",
    "../data/raw/Balancing volumes and costs_december2025.csv",
    "../data/raw/Balancing volumes and costs_januar2026.csv",
    "../data/raw/Balancing volumes and costs_februar2026.csv"
]

dfs = []
for f in months:
    df = pd.read_csv(f, sep=";", encoding="utf-8-sig", decimal=",", thousands=".", index_col=False)
    dfs.append(df)

balancing = pd.concat(dfs, ignore_index=True)
balancing["Date"] = pd.to_datetime(balancing["Date"], dayfirst=True)
balancing = balancing.rename(columns={"Date": "timestamp", "BO": "settlement_type"})
balancing = balancing.sort_values("timestamp").reset_index(drop=True)

print(balancing.shape)
print(balancing["timestamp"].min(), balancing["timestamp"].max())
balancing.to_csv("../data/curated/balancing_volumes_si_clean.csv", index=False)
print("Saved.")
```
