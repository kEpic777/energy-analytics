```python
import os
from dotenv import load_dotenv
from entsoe import EntsoePandasClient
import pandas as pd

# Load token from .env
load_dotenv()
token = os.getenv("ENTSOE_TOKEN")
print("Token loaded:", token is not None)

# Connect
client = EntsoePandasClient(api_key=token)

# Pull SI day-ahead prices — last 3 months
start = pd.Timestamp("2026-01-01", tz="Europe/Ljubljana")
end = pd.Timestamp("2026-04-01", tz="Europe/Ljubljana")

prices = client.query_day_ahead_prices("SI", start=start, end=end)
print(prices.head())
print(f"Rows: {len(prices)}")
```

```python
# Convert to DataFrame and clean
prices_df = prices.reset_index()
prices_df.columns = ["timestamp", "price_eur_mwh"]

# Save to raw and curated
prices_df.to_csv("../data/raw/entsoe_da_prices_si_2026.csv", index=False)
print(f"Saved {len(prices_df)} rows")
print(prices_df.head())
print(prices_df.dtypes)
```

```python
# Save curated version - strip timezone for simplicity
prices_df["timestamp"] = prices_df["timestamp"].dt.tz_localize(None)
prices_df.to_csv("../data/curated/da_prices_si_clean.csv", index=False)
print("Curated file saved.")
print(prices_df.head())
```
